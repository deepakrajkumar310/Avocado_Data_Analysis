### Handling categorical values

```python
first_condition = (avocado_df["region"].value_counts())/len(avocado_df)

avocado_df["region_encode"] = avocado_df["region"].apply(lambda x : first_condition[x])
```

```python
avocado_df.drop(columns=["region"],inplace=True,axis=1)
avocado_df.head()
```

```python
avocado_df = pd.get_dummies(avocado_df,columns=["type"])
avocado_df
```

```python
avocado_df.shape
```

#### Insights:

The *region* column has been handled using **frequency encoding** and the *type* column has been handled using **one hot encoding**.
